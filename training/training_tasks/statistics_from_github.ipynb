```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open('list_commits.json', 'r') as file:
    json_list = json.load(file)["list_commits"]

rating = 0
rating_list = []

for i in json_list:
    if i > 0 and i <= 3:
        j = 0
    elif i > 3:
        j = 1
    elif i <= 0:
        j = -2
    rating = rating + j
    if rating < 0:
        rating = 0
    rating_list.append(rating)

list_count = []
j = 1
for i in rating_list:
    list_count.append(j)
    j += 1

data = {'Day': list_count,
        'Unemployment_Rate': rating_list
       }

df = pd.DataFrame(data,columns=['Day','Unemployment_Rate'])
df.plot(x='Day', y='Unemployment_Rate', kind='line')
plt.show()
```
